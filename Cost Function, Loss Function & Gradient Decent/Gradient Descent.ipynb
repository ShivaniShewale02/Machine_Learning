# Gradient Descent

### What Is Gradient Descent?
- Gradient Descent is an optimization algorithm used to minimize a cost/loss function by updating model parameters in the direction of the steepest descent (negative gradient).

It is the core method used in:
- Linear Regression
- Logistic Regression
- Neural Networks
- Deep Learning



### Why Do We Need Gradient Descent?
- Most machine learning models are trained by minimizing a cost function (like MSE, cross-entropy). But we often can’t solve this minimization analytically, especially for:
- Large datasets
- Complex or non-linear models

So, we use iterative optimization → gradient descent.



### Core Idea Behind Gradient Descent
- We want to adjust parameters θ (e.g., weights) so that the cost function J(θ) becomes as small as possible.
