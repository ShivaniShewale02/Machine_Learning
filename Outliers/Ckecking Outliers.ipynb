```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Sales_data.csv")
ds.head()
```

```python
print(ds.info())
print(ds.isnull().sum())
```

```python
ds.describe()
```

### Ways to find Outliers
- Best plot to find the outliers are box plot else we can use distribution plot

```python
# Boxplot of Sales_After and the circles are the outliers as they are outside the whiskers lines
sb.boxplot(x="Sales_After", data=ds)
plt.show()  
```

```python
# displot of Sales_Before with a kernel density estimate (KDE) 
sb.displot(ds["Sales_After"], kde=True)
plt.show()
```
