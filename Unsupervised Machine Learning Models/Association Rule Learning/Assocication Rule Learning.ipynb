# Association Rule Learning

###  What is Association Rule Learning?
- Association Rule Learning is an unsupervised learning method used to identify relationships between variables/items in large datasets.
- Association rule learning is a machine Learning technique used to discover interesting relationships or patterns in large datasets.
- It is often applied to transcational data, where items are brought or used together.
