# DBSCAN Clustering Algorithm

### What is DBSCAN?
- DBSCAN is a density-based clustering algorithm that groups together points that are closely packed and marks points in low-density areas as outliers.
- Density-Based Spatial Alustering of Applications with Noise(SBSCAN)
- The clusters found by DBSCAN can be any shape, which can deal with some special cases that others methods cannot.





 Goal: Discover clusters of varying shapes and sizes in a dataset with noise.

### Intuition Behind DBSCAN
Unlike K-Means which assumes spherical clusters and requires specifying K, DBSCAN:
- Forms clusters based on density.
- Requires no prior number of clusters.
- Can detect noise/outliers.

### Python Implementation (with Visualization)

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Generate non-linear dataset
X, y = make_moons(n_samples=300, noise=0.05, random_state=0)

# DBSCAN model
model = DBSCAN(eps=0.2, min_samples=5)
labels = model.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```
