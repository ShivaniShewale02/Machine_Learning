# Model Parameters

- Model Parameters are the internal values learned from the training data by a machine learning model.
- These are the values the model uses to make predictions.
- Model parameters are cofiguration varibales that are internal to the models and a model learns them on its own.

### Example

- for example linearn model there m and c are the parameters which helps us to predict the final output.

### Difference: Parameters vs Hyperparameters

| Aspect             | **Parameters**                    | **Hyperparameters**                      |
| ------------------ | --------------------------------- | ---------------------------------------- |
| Defined by         | Model during training             | You (before training)                    |
| Examples           | Weights, biases                   | Learning rate, max\_depth, n\_estimators |
| Learned from data? | ✅ Yes                             | ❌ No                                     |
| Tuned with?        | Gradient Descent, Backpropagation | Grid Search, Random Search, Optuna, etc. |


### Examples of Model Parameters by Algorithm

| Algorithm               | Model Parameters                     |
| ----------------------- | ------------------------------------ |
| **Linear Regression**   | Weights $w_i$, bias $b$              |
| **Logistic Regression** | Weights $w_i$, bias $b$              |
| **Decision Tree**       | Node splits, split feature/threshold |
| **KNN**                 | ❌ No parameters — lazy learner       |
| **SVM**                 | Support vectors, coefficients        |
| **Neural Network**      | Weights and biases in all layers    
| **Naive Bayes**         | Probabilities (P(feature class), P(class)) |
