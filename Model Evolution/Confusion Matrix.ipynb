# Confusion Matrix

### 1. What Is a Confusion Matrix?
- A confusion matrix is a summary table that shows how well a classification model performs by comparing predicted labels vs actual labels.
- It breaks down true positives, false positives, true negatives, and false negatives.
-  A confusion matrix is a simple and useful tool for understanding the performance of a classification model, like one used in machine learning or stastistics.
-  It helps you evaluate how well your model is doing in categorizing things correctly.
-  It is also known as the error matrix.
-  The matrix consists of predictions result in a summarized form, which has a number of correst predictions and incorrect predictions.

### onfusion Matrix for Binary Classification
- üí° Example:
Let‚Äôs say your model predicts whether a patient has cancer (1) or not (0).

|                          | **Predicted: 1 (Positive)** | **Predicted: 0 (Negative)** |
| ------------------------ | --------------------------- | --------------------------- |
| **Actual: 1 (Positive)** | ‚úÖ True Positive (TP)        | ‚ùå False Negative (FN)       |
| **Actual: 0 (Negative)** | ‚ùå False Positive (FP)       | ‚úÖ True Negative (TN)        |




- it is created by y original and y predicted
- 0 is false and 1 is true.

### Key Metrics Derived from Confusion Matrix
##### Accuracy = TN + TP / TN + TP + FP + FN


##### 2. Precision (Positive Predictive Value):
- Out of all predicted positives, how many were correct?



##### 3. Recall (Sensitivity, True Positive Rate):
- Out of all actual positives, how many did we correctly predict?



##### 4. F1 Score (Harmonic mean of precision & recall):


##### 5. Specificity (True Negative Rate):


- Make sure that false negatice should be zero and the least if that is more then the model is not well performing.
- Recall: it helps us to measure how many positive samples were correctly classified by the ML model.
- recall is inversly proprotional to False Negative.
- Precision: It helps us to measure the ability to classify positive samples in the model.
- F1 Score: It is the harmonic mean of precision and recall. it takes bith false positive and false negative into account. Therefore, it performs well on an imbalanced dataset.




| Metric      | Formula                         | Purpose                         |
| ----------- | ------------------------------- | ------------------------------- |
| Accuracy    | $\frac{TP + TN}{Total}$         | Overall correctness             |
| Precision   | $\frac{TP}{TP + FP}$            | Correctness of positive guesses |
| Recall      | $\frac{TP}{TP + FN}$            | Ability to find all positives   |
| F1 Score    | $2 \cdot \frac{P \cdot R}{P+R}$ | Balance of P and R              |
| Specificity | $\frac{TN}{TN + FP}$            | Catching true negatives         |


```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# True and predicted labels
y_true = [0, 0, 1, 1, 0, 1, 0, 1]
y_pred = [0, 1, 1, 1, 0, 0, 0, 1]

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n", classification_report(y_true, y_pred))

# Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
