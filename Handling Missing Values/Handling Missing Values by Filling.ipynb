# For Catogerical Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Sales_data.csv")
ds.head(10)
```

```python
ds.isnull().sum()
```

```python
ds.fillna(10)    # use to fill the data at the empty place (here we have directly filled 10 as data)
```

```python
ds.info()     # use to get the info of the data
```

```python
ds.fillna(method= "bfill")   # use to fill the data from the data of row bellow NA row
```

```python
ds.fillna(method= "ffill")    # use to fill the data from the data of row above NA row
```

```python
ds.fillna(method= "ffill", axis = 1)  # use to fill the data from the data one before the NA data blovk (horizontal data filling with axis = 1)
```

```python
ds.fillna(method= "bfill", axis = 1)  # use to fill the data from the data one after the NA data blovk (horizontal data filling with axis = 1)
```

```python
# Mode Data Filling
ds['Customer_Satisfaction_After'].mode()[0]   # taking mode of the given column 
ds['Customer_Satisfaction_After'].fillna(ds['Customer_Satisfaction_After'].mode()[0], inplace = True )    # here filling the value of mode at the place of empty data
```

```python
ds.select_dtypes(include='object')    # use to get the specific type of data from the dataset
```

```python
ds.select_dtypes(include='object').isnull().sum()
```

```python
ds.select_dtypes(include="object").columns
```

```python
for i in ds.select_dtypes(include="object").columns:
    ds[i].fillna(ds[i].mode()[0], inplace = True )
```

```python
ds.select_dtypes(include="object").isnull()
```
