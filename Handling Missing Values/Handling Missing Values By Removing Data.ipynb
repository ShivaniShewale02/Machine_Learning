```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Sales_data.csv")   # Reading the data from the csv file
ds.head(5)   # use to print the limited top portion of the data 
```

```python
ds.shape    # use to get the no. of rows and columns in the dataset ( here 1000 rows and 7 columns )
```

```python
print(ds.info())    # Overview of missing values
ds.isnull().sum()   # use to get the numberic form of the no. of null values in the data or specific in tables
```

```python
sns.heatmap(ds.isnull())
plt.show()
```

# Handling missing data 
- by removing the column itself
- by removing the row where the data is missing  

```python
# Removing Column
ds.drop(columns=["Customer_Segment"], inplace= True)  # here we used it to remove the given column and save it in the same dataframe (if i do not use the inplace = True then it will create me a new dataframe)
```

```python
print(ds.shape)
ds.isnull().sum()
```

```python
# Removing Rows with missing values or content 
ds.dropna(inplace= True)  # use to remove all the rows with missing value at once with this function
```

```python
sns.heatmap(ds.isnull())
plt.show()
```
