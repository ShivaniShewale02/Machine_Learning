# Feature Scaling Using Normalization

- Normalization (also called Min-Max Scaling) is a feature scaling technique that transforms data to a fixed range (typically 0 to 1). 
- This ensures that all features contribute equally to a machine learning model.
- Formula for Normalization (Min-Max Scaling):
- X′ = X − Xmin / Xmax − Xmin
- Where:
- X = Original feature value
- Xmin = Minimum value in the feature
- Xmax= Maximum value in the feature
- X′= Transformed value in the range [0,1]

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
ds = pd.read_csv('Sales_data.csv')
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds.fillna({'Sales_Before': ds['Sales_Before'].mean(), 'Sales_After': ds['Sales_After'].mean(), 'Customer_Satisfaction_Before': ds['Customer_Satisfaction_Before'].mean(), 'Customer_Satisfaction_After': ds['Customer_Satisfaction_After'].mean()}, inplace =True)
ds.isnull().sum()
```

```python
sns.displot(ds['Sales_After'], kde=True)
plt.plot()
sns.displot(ds['Sales_Before'], kde=True)   
plt.plot()
sns.displot(ds['Customer_Satisfaction_Before'], kde=True)
plt.plot()
sns.displot(ds['Customer_Satisfaction_After'], kde=True)   
plt.plot()
```

```python
# Apply MinMaxScaler Function to normalize the data
scaler = MinMaxScaler()
ds[['Sales_Before1', 'Sales_After', 'Customer_Satisfaction_Before', 'Customer_Satisfaction_After']] = scaler.fit_transform(ds[['Sales_Before', 'Sales_After', 'Customer_Satisfaction_Before', 'Customer_Satisfaction_After']])
ds.describe()
```

```python
# For Before_sales (just to check the distribution after standardization)
plt.subplot(1, 2, 1)
plt.title('Sales Before Standardization')
sns.histplot(ds['Sales_Before'], kde=True)

plt.subplot(1, 2, 2)
plt.title('Sales After Standardization')
sns.histplot(ds['Sales_Before1'], kde=True)

plt.show()
```

```python
sns.displot(ds['Sales_Before'], kde=True)   
plt.plot()

sns.displot(ds['Customer_Satisfaction_Before'], kde=True)
plt.plot()

sns.displot(ds['Customer_Satisfaction_After'], kde=True)   
plt.plot()
```
